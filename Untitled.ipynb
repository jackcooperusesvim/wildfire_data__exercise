```python
uniques: Dict[Tuple[str,str],List[str]] = {}

for col in encodable_cols:
    uniques[col] = list(pge_24_df[col].unique())


for col in encodable_cols:
    new_col = (col[0],col[1]+"_encoded")
    pge_24_df[new_col] = pge_24_df[col].astype("category").cat.codes
```

```python
def trim_na(df, per_col_limit):
    for col in df:
        if df[col].count() >= per_col_limit:
            df = df[~df[col].isna()]
    return df
        
pge_24_df = trim_na(pge_24_df,500)
```
